```python
import openpyxl
from datetime import datetime # datetimeモジュールをインポート
#　新規ワークブックを作成
wb = openpyxl.Workbook()
#　アクティブなシートを変数「sheet」に格納
ws = wb.active

#　B2のセルに「請求書」を入力
ws['B2'] = '請求書'

#　B4～B7まで固定データを入力,F4に「No.」F5に「日付」を入力
ws['B4'] = '株式会社ABC'
ws['B5'] = '〒101-0022 東京都千代田区神田練塀町300'
ws['B6'] = 'TEL:03-1234-5678 FAX:03-1234-5678'
ws['B7'] = '担当者名:鈴木一郎 様'
ws['F4'] = 'No.'
ws['F5'] = '日付'

# G4に請求書番号を設定する
invoice_number = '0001'

# G4に表示する
ws['G4'] = invoice_number

# B5からE5までセルを結合する
ws.merge_cells('B5:E5')

# B6からE6までセルを結合する
ws.merge_cells('B6:E6')

# B7からD7までセルを結合する
ws.merge_cells('B7:D7')

# G5からH5までセルを結合する
ws.merge_cells('G5:H5')

# 現在の年月日を取得
current_date = datetime.now().strftime("%Y/%m/%d")

# G5に年月日を表示する
ws['G5'] = current_date

# B10～E10まで「商品名」「数量」「単価」「金額」を入力
ws['B10'] = '商品名'
ws['C10'] = '数量'
ws['D10'] = '単価'
ws['E10'] = '金額'

# 商品、数量、単価のリストを作成
products = [
     ['商品A', 2, 10000],
     ['商品B', 1, 15000]
]


# 書き込みを開始する行を設定
row = 11

# リストに対してループ処理をする
for product in products:
    # 現在の行にリストの内容を入力
    ws[f'B{row}'] = product[0]
    ws[f'C{row}'] = product[1]
    ws[f'D{row}'] = product[2]
    ws[f'E{row}'] = product[1] * product[2]
    # 現在の行を1つ下に移動する
    row +=1

# B15～B17まで「小計」「消費税」「合計」を入力
ws['B15'] = '合計'
ws['B16'] = '消費税'
ws['B17'] = '税込合計'

# E15,E16,E17に計算して表示する。
ws['E15'] = f'=sum(E11:E{row-1})' # 範囲はE11からE14
ws['E16'] = f'=E15*0.1'
ws['E17'] = f'=E15+E16'

wb.save('請求書_20241013.xlsx')
```
